```python
for i in range(len(sent)):
    words=nltk.word_tokenize(sent[i])
    ###in this we tokenise the sentence to word using sent iteration
    words= [lema.lemmatize(word) for word in words  if word not in set(stopwords.words('english'))]
    ###after we lematise and we remove stop words
    sent[i]=' '.join(words)
    ##finally we join
```
