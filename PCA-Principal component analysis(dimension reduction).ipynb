```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
df.keys()
```

```python
df1=pd.DataFrame(df['data'],columns=df.feature_names)
```

```python
df1
```

```python
print(df['DESCR'])
```

### we have 30 dimensions let's decrease those....

### first we scale the data........ 

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(df1)
```

```python
df_sc=scalar.transform(df1)
df_sc
pd.DataFrame(df_sc,columns=df.feature_names)
```

### after scaling the data we apply the P.C.A

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(df_sc)
```

```python
df_sc_pca=pca.transform(df_sc)
```

```python
df_sc_pca
```

```python
pd.DataFrame(df_sc_pca,columns=['pc1','pc2'])
```

```python
df1.shape
```

```python
df_sc_pca.shape
```

 ###  we know that df_sc_pca[ rows:,columns ]

```python
plt.scatter(df_sc_pca[:,0],df_sc_pca[:,1],c=df['target'], cmap='plasma')
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.figure(figsize=(15,20))
```

### plotting the reducede dimensions in dataframe

### 1.standard scaling
2.pca(fit then transform)

```python
from sklearn.model_selection import train_test_split
```

```python
end_df=pd.DataFrame(df_sc_pca,columns=['d1','d2'])
end_df
```

```python
end_df['target']=df.target
```

```python
end_df
```

```python
x=end_df.drop('target',axis=1)
y=end_df['target']
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression( random_state=33)
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
model.score(xtrain,ytrain)
```

### it gives us high accuracy and standard accuracy

```python
model.predict_proba(xtest)
```

```python
y_pred=model.predict(xtest)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_auc_score
```

```python
cm=confusion_matrix(ytest,y_pred)
sns.heatmap(cm,annot=True)
```
